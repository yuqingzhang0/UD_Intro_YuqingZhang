```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options
import time
from bs4 import BeautifulSoup
import os
from time import sleep

chromedriver = r"C:\Users\zyq\ladybug_tools\python\Scripts\chromedriver.exe"

#headless mode
options = webdriver.ChromeOptions()
options.add_argument('--headless=new')

driver = webdriver.Chrome(options=options)
    
url = 'https://www.nintendo.co.uk/'

driver = webdriver.Chrome()
driver.get(url)

# Locate the "Games-Nintendo Switch Games" button on the webpage and click it to navigate to the next page
driver.find_element("xpath","//*[@id='mainNav_games']/div[2]").click()
driver.find_element("xpath","//*[@id='scndlvl_games_NintendoSwitchgames']").click()

# Create lists
games_list = []

# Cancel 'Nintendo Switch' button
driver.find_element("xpath","//*[@id='VA_MainSearch']/div[2]/div/div/div[1]/div/div[1]/div/ul/li[1]/label").click()

# search 'ZELDA'
search_query = driver.find_element("xpath","//*[@id='VA_MainSearch']/div[1]/div[1]/form/input")
search_query.send_keys('zelda')
sleep(2)
search_query.send_keys(Keys.RETURN)
sleep(5)

# Get results for the first page
results_id = driver.find_elements('class name','search-app')
results1 = results_id[0].find_elements('class name','searchresult_row.page-list-group-item.col-xs-12')
print(len(results1))
    
#Create loop to itterate on the list
for i in range(len(results1)):
    
    # Extracting title
    title1 = results1[i].find_element("class name","page-title-text").text
    sleep(5)
          
    # Extracting links
#     link1 = results1[i].find_elements("xpath","//div[@class='col-xs-12']/a[position() = 1 and @href]")
#     print(len(link1))
#     for i in range(len(link1)):
#         link1 = link1(i = i[:-2])
#         link = link1[i].get_property('href')
#         print (link)
#     sleep(5)
        
    # Add items to the respective lists
    games_list.append(title1)
    links_list1 = results1[0].find_elements("xpath","//div[@class='col-xs-12']/a[position() = 1 and @href]")[:-2]
    links_list1 = [l.get_property('href') for l in links_list1]
    print(games_list)
    
    
    sleep(5)
    
print(links_list1)
sleep(8)

# Press the next button
driver.find_element("xpath","//*[@id='VA_MainSearch']/div[2]/div/div/div[3]/div/button[3]").click()
sleep(8)

results2_id = driver.find_elements('id','VA_MainSearch')
# Get results for the second page
results2 = results2_id[0].find_elements('class name','searchresult_row.page-list-group-item.col-xs-12')
print(len(results2))

    
for i in range(len(results2)):

    # Extracting title
    title2 = results2[i].find_element("class name","page-title-text").text
    print (title2)
    sleep(5)
          
    # Extracting links
#     link2 = results2[i].find_element("xpath","//div[@class='col-xs-12']/a[position() = 1 and @href]").get_property('href')
#     print (link2)
#     sleep(5)
        
    # Add items to the respective lists
    games_list.append(title2)
    links_list2 = results2[0].find_elements("xpath","//div[@class='col-xs-12']/a[position() = 1 and @href]")[:-2]
    links_list2 = [l.get_property('href') for l in links_list2]
    print(games_list)
    sleep(5)
links_list = (links_list1) + (links_list2)
print(links_list)
driver.quit()

```

```python
# Import the library
import csv
import requests
import os

data = zip(games_list,links_list)

with open('nintendogames.csv','w',newline = '',encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    for row in data:
        writer.writerow(row)
        
#Write only links to a csv
with open('onlygameslinks.csv','w',newline = '',encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    for link in links_list:
        writer.writerow([link])

# Import a new library & open gallery
import requests
with open("onlygameslinks.csv",'r',encoding = 'utf-8') as file:
    csv_reader = csv.reader(file)
    for j, url in enumerate(csv_reader): 
        #open each link
        driver = webdriver.Chrome()
        print (url)
        driver.get(url[j])
        path = games_list[j].replace(':', '_')
        if not os.path.exists(path):
            os.mkdir(path)
        sleep(8)
        
        #click on the gallery button
        gallery_link = driver.find_element("xpath","//*[@id='gamePageNav']/nav/div[2]/div[2]/div/ul")
        sleep(5)
        gallery_link.find_element("xpath","//a[position() = 1 and @href = '#Gallery']").click()
        sleep(3)
        image_link_upper = driver.find_elements("class name","mediagallery-image-overlay")
        a = 0
        for link_upper in image_link_upper:
            image_final = link_upper.find_element('xpath', '../img').get_attribute('src')
            print(image_final)
            suffix = image_final.split('.')[-1]
            a += 1
            print("Downloading file:",a)
            response = requests.get(image_final)
            new_name = "image"+str(a)+"."+suffix
            filepath = os.path.join(filepath,new_name)
            with open(os.path.join(path,new_name),"wb") as f:
                f.write(response.content)
                sleep(5)
    
        break
        
# END
```
